2024

```python
file_path = 'Past_Top500_HPCs/TOP500_202406.xlsx'
data = pd.read_excel(file_path, sheet_name='63')

site_data = data[['Rank', 'Name','Country','Rmax [TFlop/s]','Power (kW)']].head(500)
dropna=site_data.dropna(subset=['Power (kW)'])
power=list(dropna['Power (kW)'])
full_energy=[power[i]*31536000 for i in range (len(dropna['Power (kW)']))]
avg2024=(sum(power)/len(dropna['Power (kW)']))
```

linear regression


```python


years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024]).reshape(-1, 1)
power_consumption = np.array([1750.5487234042553, 1756.6032057416269, 1727.632275132275, 
                              1900.5513812154695, 1782.6515706806283, 1813.9878191489363, 
                              2090.7377319587627])


model = LinearRegression()
model.fit(years, power_consumption)


predicted_values = model.predict(years)


slope = model.coef_[0]
intercept = model.intercept_

(slope, intercept, predicted_values)




plt.plot(years, power_consumption, marker='o', linestyle='', label='Original Data')
plt.plot(years, predicted_values, color='orange', label='Linear Regression Line')
plt.xlabel('Year')
plt.ylabel('Average Power Consumption of HPCs (kW)')
plt.title('Average Power Consumption of HPCs (kW) from 2018 to 2024')
plt.legend()
plt.show()

```
