```python
file_path = 'Past_Top500_HPCs/TOP500_202406.xlsx'
data = pd.read_excel(file_path, sheet_name='63')

site_data = data[['Rank', 'Name','Country','Rmax [TFlop/s]','Power (kW)']].head(500)
dropna=site_data.dropna(subset=['Power (kW)'])
power=list(dropna['Power (kW)'])
full_energy=[power[i]*31536000 for i in range (len(dropna['Power (kW)']))]
avg2024=(sum(power)/len(dropna['Power (kW)']))
```

linear regression


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Define the data for years and power consumption
years = np.array([2018, 2019, 2020, 2021, 2022, 2023, 2024]).reshape(-1, 1)
values = np.array([
    1750.5487234042553, 1756.6032057416269, 1727.632275132275, 
    1900.5513812154695, 1782.6515706806283, 1813.9878191489363, 
    2090.7377319587627
])

# Initialize and fit the linear regression model
linear_model = LinearRegression()
linear_model.fit(years, values)

# Predict future values for 2025-2028
future_years = np.array([2025, 2026, 2027, 2028]).reshape(-1, 1)
predicted_values = linear_model.predict(future_years)

# Combine original and predicted years and values for plotting
extended_years = np.concatenate((years.flatten(), future_years.flatten()))
extended_values = np.concatenate((values, predicted_values))

# Generate predictions for all years (both original and future) for the regression line
all_years = extended_years.reshape(-1, 1)
all_predicted_values = linear_model.predict(all_years)

# Plot original data, predictions, and the regression line
plt.figure(figsize=(10, 6))
plt.plot(years.flatten(), values, 'o', color='orange', label='Original Data')  # Original data in orange
plt.plot(future_years.flatten(), predicted_values, 'o', color='blue', label='Predicted Data')  # Predicted data in blue
plt.plot(extended_years, all_predicted_values, linestyle='-', color='green', label='Linear Regression Line')  # Regression line in green

# Set x-axis labels to display all years
plt.xticks(extended_years)

plt.xlabel('Year')
plt.xlabel('Year')
plt.ylabel('Average Power Consumption of HPCs (kW)')
plt.title('Average Power Consumption of HPCs (kW) from 2018 to 2024')
plt.legend()
plt.savefig('predictions/Prediction of Average Power Consumption of HPCs (kW) from 2018 to 2028 with Linear Regression')
plt.show()

# Retrieve and print the slope and intercept of the regression line
slope = linear_model.coef_[0]
intercept = linear_model.intercept_
print("Slope:", slope)
print("Intercept:", intercept)
```
